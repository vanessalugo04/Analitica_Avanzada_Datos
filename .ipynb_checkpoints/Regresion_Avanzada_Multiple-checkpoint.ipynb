# Regresión Avanzada Múltiple

Se importan las librerias necesarias para manipulación de datos, creación de gráficas y entrenamiento
del modelo de regressión avanzada múltiple.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

Conjunto de datos

```python
X = np.array([
    [73, 2, 25],
    [81, 2, 30],
    [61, 1, 60],
    [48, 1, 55],
    [45, 1, 50],
    [50, 2, 110],
    [100, 3, 70],
    [60, 2, 20],
    [40, 1, 90],
    [105, 3, 15]
])

Y = np.array([14.5, 12.5, 9.5, 6, 7, 8.5, 15, 14.5, 7.8, 25])
```

Entrenamiento

```python
modelo = LinearRegression()
```

```python
modelo.fit(X, Y)
```

Predicciones

Calculo de los coeficientes

```python
b0 = modelo.intercept_
b1_b2_b3 = modelo.coef_
print(f'b0 = {b0}')
print(f'b1, b2, b3= {b1_b2_b3}')
```

```python
Y_pred = modelo.predict(X)
print(Y_pred)
```

Cálculo de coeficiente de determinación

```python
r2 = r2_score(Y, Y_pred)
print(f'El coeficiente de determinación R^2 es: {r2:.4f}')
```

Gráfica de resultados

```python
plt.figure(figsize=(8, 6))
plt.scatter(Y, Y, color = 'blue', label='Datos originales') 
plt.scatter(Y, Y_pred, color = 'red', label = 'Predicciones')
plt.plot([min(Y), max(Y)], [min(Y_pred), max(Y_pred)], color = 'black', linestyle='-', label = 'Regresión')
plt.xlabel('Renta real')
plt.ylabel('Predicción de renta')
plt.legend()
plt.grid(True)
plt.show()
```

Calculo de errores

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

error_aboluto_medio = mean_absolute_error(Y, Y_pred)
error_cuadratico_medio = mean_squared_error(Y, Y_pred)
raiz_error_cuadratico_medio = np.sqrt(error_cuadratico_medio)

print(f'Error absoluto medio = {error_aboluto_medio}')
print(f'Error cuadratico medio = {error_cuadratico_medio}')
print(f'Raiz del error absoluto medio = {raiz_error_cuadratico_medio}')
```
