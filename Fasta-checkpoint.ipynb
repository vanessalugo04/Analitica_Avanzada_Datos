## Importar librerías
Se usa from fastai.tabular.all import * para crear data loaders. 

```python
from fastai.tabular.all import *
import pandas as pd
```

## Carga de datos
Se decarga el conjunto de datos con under_data aue descarga y extrae el URL por defecto en subdirecorios de fastai y devuelve la ruta

```python
path = untar_data(URLs.ADULT_SAMPLE) #Se descargan los datos
```

```python
df = pd.read_csv(path/'adult.csv') #Se convierte a Data Frame
```

```python
df.head() #Se muestran los primeros 5 registros
```

## Variables
Se asignan las variables dependientes, categoricas y numericas. La variable dependiente corresponde al objetivo, es decir, el salario de las personas. Las variables catategóricas son aquellas que representan características, mientras que las variables numéricas corresponden a valores tipo entero o flotantes. 

```python
var_dep = 'salary' #Se asigna variable dependiente
```

```python
var_categoricas = ['workclass' , 'education' , 'marital-status' , 'occupation', 'relationship',
                   'race', 'sex', 'native-country'] #Variables categóricas
var_numericas = ['age', 'fnlwgt', 'education-num', 'capital-loss', 'hours-per-week'] #Variables numéricas
```

## Preprocesadores
Categorify: Convierte variables categóricas a numéricas.    
FillMissing: Rellena valores faltantes con la mediana de los valores existentes.   
Normalize: Normaliza las variables continuas, de manera que se resta la media y se divide por la desviación estándar.   

```python
procs = [Categorify, FillMissing, Normalize] #Se declara la lista de preprocesadores
```

## Data Loaders
FastAI da una alternativa a DataLoader de PyTorch. Recibe como parámtros:    
cat_names: Son los nombres de las variables categóricas.    
cont_names: Son los nombres de las variables continuas.    
y_names: Son los nombres de las variables dependientes.   
y_block: Es el TransformBlockque se utilizará para el objetivo.    
bs: Es el tamaño del lote (int)

```python
df = df.copy()
dls = TabularDataLoaders.from_df(df, path = path, procs = procs,
                                 cat_names = var_categoricas, cont_names = var_numericas,
                                 y_names = var_dep, y_block = CategoryBlock(), bs = 64)
```

## Creación del modelo
Se usar el método tabular_learner para entrenar el modelo, redibe como parámetros los datos, es decir, data loaders, se define el tamaño de la red nauronal, en este caso son dos capas con 200 nauronas en la primer capa y 100 neuronas en la segunda capa, y se define la métrica de evaluación, que es la exactitud del modelo. 

```python
learn = tabular_learner(dls, layers = [200, 100], metrics = accuracy) #Se establecen parámetros del modelo
```

## Arquitectura del modelo

```python
learn.model #Se visualiza la estructura interna del modelo
```

## Entrenamiento del modelo
Se indica el orden en que se va entrenar el modelo, de manera que lo va haciendo por épocas, en este caso son 3.

```python
learn.fine_tune(3)  #Se entrena el modelo
```

## Evaluación del modelo
Se observan los resultados del modelo, de manera que se pueden comparar los valores reales de la variable dependiente de sus valores predichos.

```python
learn.show_results() 
```

## Matriz de confusión
Se crea y se muestra la matriz de confusión para evaluar la clasificación del modelo.

```python
from fastai.interpret import ClassificationInterpretation #Se importa la clase
interp = ClassificationInterpretation.from_learner(learn) #Calcula la matriz de confusón
interp.plot_confusion_matrix() #Imprime la matriz de confusión
```

## Conclusión
Se observa que la predicción de los salarios fue érronea en bastantes casos al observar los resultados de comparación de variable dependiente real vs la predicha, y esto se visualiza en la matriz de confusión, pues el modelo no aprendió a clasificar de manera correcta debido a que tiene 1636 falsos positivos y ningún verdadero negativo. Esto puede ser al desbalance de clases, o a los parámetros usados al entrenar.
